--Created a function to remove two decimal points

def clean_price(value):
    if pd.isna(value):
        return np.nan
    # Remove all dots except the last one
    parts = value.split('.')
    if len(parts) > 2:
        # join everything except last part as thousand-sep
        value = ''.join(parts[:-1]) + '.' + parts[-1]
    # Replace comma with dot (European decimal to standard)
    # value = value.replace(',', '.')
    try:
        return float(value)
    except:
        return np.nan
